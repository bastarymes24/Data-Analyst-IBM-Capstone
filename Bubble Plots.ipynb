### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
df_plot = df[["Age", "SOPartFreq", "JobSat"]].copy()
df_plot["Age"] = pd.to_numeric(df_plot["Age"], errors="coerce")
df_plot["JobSat"] = pd.to_numeric(df_plot["JobSat"], errors="coerce")
df_plot = df_plot.dropna()

plt.figure(figsize=(8,6))
plt.scatter(
    df_plot["Age"],
    df_plot["SOPartFreq"].astype("category").cat.codes,
    s=df_plot["JobSat"] * 20,
    alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Participation Frequency (Encoded)")
plt.title("Age vs Participation Frequency (Bubble = Job Satisfaction)")
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
df_plot = df[["ConvertedCompYearly", "JobSat", "Age"]].copy()
df_plot = df_plot.apply(pd.to_numeric, errors="coerce")
df_plot = df_plot.dropna()

plt.figure(figsize=(8,6))
plt.scatter(
    df_plot["ConvertedCompYearly"],
    df_plot["JobSat"],
    s=df_plot["Age"] * 2,
    alpha=0.5)
plt.xlabel("ConvertedCompYearly")
plt.ylabel("Job Satisfaction")
plt.title("Compensation vs Job Satisfaction (Bubble = Age)")
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here
df_lang = df[["Age", "LanguageHaveWorkedWith"]].dropna()
df_lang["Age"] = pd.to_numeric(df_lang["Age"], errors="coerce")
df_lang = df_lang.dropna()
lang_counts = (
    df_lang.assign(Language=df_lang["LanguageHaveWorkedWith"].str.split(";"))
           .explode("Language")
           .groupby(["Age", "Language"])
           .size()
           .reset_index(name="Count"))

plt.figure(figsize=(10,6))
plt.scatter(
    lang_counts["Age"],
    lang_counts["Language"].astype("category").cat.codes,
    s=lang_counts["Count"] * 5,
    alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Programming Language (Encoded)")
plt.title("Technology Popularity by Age")
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
df_db = df[["DatabaseWantToWorkWith", "JobSat"]].dropna()
df_db["JobSat"] = pd.to_numeric(df_db["JobSat"], errors="coerce")
df_db = df_db.dropna()
db_counts = (
    df_db.assign(Database=df_db["DatabaseWantToWorkWith"].str.split(";"))
         .explode("Database")
         .groupby("Database")
         .agg({"JobSat":"mean", "Database":"count"})
         .rename(columns={"Database":"Count"})
         .reset_index())

plt.figure(figsize=(10,6))
plt.scatter(
    db_counts["JobSat"],
    db_counts["Database"].astype("category").cat.codes,
    s=db_counts["Count"] * 10,
    alpha=0.6)
plt.xlabel("Average Job Satisfaction")
plt.ylabel("Database (Encoded)")
plt.title("Preferred Databases vs Job Satisfaction")
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here
df_dev = df[["DevType", "ConvertedCompYearly", "JobSat"]].copy()
df_dev["ConvertedCompYearly"] = pd.to_numeric(df_dev["ConvertedCompYearly"], errors="coerce")
df_dev["JobSat"] = pd.to_numeric(df_dev["JobSat"], errors="coerce")
df_dev = df_dev.dropna(subset=["ConvertedCompYearly", "JobSat", "DevType"])
df_dev = df_dev.assign(Role=df_dev["DevType"].str.split(";")).explode("Role")
dev_counts = (
    df_dev.groupby("Role")
          .agg({
              "ConvertedCompYearly": "mean",
              "JobSat": "mean"
          })
          .reset_index())

plt.figure(figsize=(10,6))
plt.scatter(
    dev_counts["ConvertedCompYearly"],
    dev_counts["Role"].astype("category").cat.codes,
    s=dev_counts["JobSat"] * 20,
    alpha=0.6
)
plt.xlabel("Average Compensation")
plt.ylabel("Developer Role (Encoded)")
plt.title("Compensation Across Developer Roles")
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
df_tools = df[["Age", "NEWCollabToolsHaveWorkedWith"]].dropna()
df_tools["Age"] = pd.to_numeric(df_tools["Age"], errors="coerce")
df_tools = df_tools.dropna()
tool_counts = (
    df_tools.assign(Tool=df_tools["NEWCollabToolsHaveWorkedWith"].str.split(";"))
            .explode("Tool")
            .groupby(["Age", "Tool"])
            .size()
            .reset_index(name="Count"))

plt.figure(figsize=(10,6))
plt.scatter(
    tool_counts["Age"],
    tool_counts["Tool"].astype("category").cat.codes,
    s=tool_counts["Count"] * 5,
    alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Collaboration Tool (Encoded)")
plt.title("Collaboration Tools by Age")
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here
df_web = df[["WebframeWantToWorkWith", "JobSat"]].dropna()
df_web["JobSat"] = pd.to_numeric(df_web["JobSat"], errors="coerce")
df_web = df_web.dropna()
web_counts = (
    df_web.assign(Web=df_web["WebframeWantToWorkWith"].str.split(";"))
          .explode("Web")
          .groupby("Web")
          .agg({"JobSat":"mean", "Web":"count"})
          .rename(columns={"Web":"Count"})
          .reset_index())

plt.figure(figsize=(10,6))
plt.scatter(
    web_counts["JobSat"],
    web_counts["Web"].astype("category").cat.codes,
    s=web_counts["Count"] * 10,
    alpha=0.6)
plt.xlabel("Average Job Satisfaction")
plt.ylabel("Web Framework (Encoded)")
plt.title("Preferred Web Frameworks vs Job Satisfaction")
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
##Write your code here
df_admire = df[["Country", "LanguageAdmired"]].dropna()
admire_counts = (
    df_admire.assign(Language=df_admire["LanguageAdmired"].str.split(";"))
             .explode("Language")
             .groupby(["Country", "Language"])
             .size()
             .reset_index(name="Count"))

plt.figure(figsize=(10,6))
plt.scatter(
    admire_counts["Country"].astype("category").cat.codes,
    admire_counts["Language"].astype("category").cat.codes,
    s=admire_counts["Count"],
    alpha=0.5)
plt.xlabel("Country (Encoded)")
plt.ylabel("Admired Technology (Encoded)")
plt.title("Admired Technologies Across Countries")
plt.show()
```
